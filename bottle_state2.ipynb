```python
import numpy as np
import os
import PIL
import PIL.Image
import tensorflow as tf
import tensorflow_datasets as tfds
import pathlib
```

```python
batch_size = 32
img_height = 224
img_width = 224
```

절취선

```python
bottle_dir = "/content/drive/MyDrive/project2/data_test_class/bottle"
bottle_dir = pathlib.Path(bottle_dir)
```

```python
bottle_feature = tf.keras.preprocessing.image_dataset_from_directory(
  bottle_dir,
  labels=None,
  seed=12,
  shuffle=False,
  image_size=(img_height, img_width),
  batch_size=batch_size)
```

```python
image_count = len(list(bottle_dir.glob('*.png')))
print(image_count)
```

```python
model_bottle = tf.keras.models.load_model ("/content/drive/MyDrive/h5파일모음/model_bottle.h5")
```

```python
bottle_predict = model_bottle.predict(bottle_feature)
```

```python
bottle_list = []


for i in range(len(bottle_predict)):
    id = np.argmax(bottle_predict[i])
    bottle_list.append(id)
```

```python
# bottle_classes = ['bottle-broken_large', 'bottle-broken_small', 'bottle-contamination', 'bottle-good']
bottle_classes = os.listdir("/content/drive/MyDrive/project2/data_train_state/bottle")
```

```python
bottle_classes.sort()
bottle_classes
```

```python
bottle_list2 = []
for i in range(len(bottle_list)):
    bottle_list2.append(bottle_classes[bottle_list[i]])

print(bottle_list2)
```

```python
filelist = os.listdir("/content/drive/MyDrive/project2/data_test_class/bottle")
print(filelist)
filelist.sort()
print(filelist)
```

```python
import pandas as pd
classname = "bottle-"

for i in range(len(bottle_list2)):
    bottle_list2[i] = classname + bottle_list2[i]

print(bottle_list2)

```

```python
bottle_df = pd.DataFrame({"filename" : filelist,
                         "label":bottle_list2 })

bottle_df.head()

# 이렇게 데이터프레임 만들어서 저장하고, 나머지 다 합쳐서 filename기준으로 sort쓴다음 filename컬럼 제거해서 제출형식으로 만들 예정
```
